# <font color='#33FFFF '>Introduction :</font>

Twitter est un site Internet,Inventé par l'américain Jack Dorsey en mars 2006,qui permet de diffuser gratuitement des minimessages vers des ordinateurs ou des téléphones portables. Ces textes qu'on appelle des « tweets » peuvent contenir jusqu'à  140 signes alphabétiques, (c'est-à-dire 140 lettres ou espaces). Pas 1 de plus !<br>
#### <font color='red'> Comment ça marche ? </font> <br>
Pour pouvoir envoyer un message sur Twitter, il faut d'abord s'inscrire sur Internet. C'est gratuit. Une fois cette étape réalisée, on peut décider d'écrire soi-même des tweets qui sont diffusés et lus par d'autres membres de la communauté Twitter qui ont envie de « suivre » ce que l'on raconte. On appelle ces personnes les « followers » ceux qui suivent vos messages. On peut aussi suivre des messages envoyés par d'autres membres de Twitter pour découvrir leur avis, on aura alors des « followings ».<br>
#### <font color='red'> À quoi sert Twitter ? </font> <br>
C'est un site qui permet de diffuser et de lire rapidement des informations courtes sur des sujets qui t'intéressent : un avis sur un match sportif, une actualité sur un chanteur, une critique sur un film récent… Twitter ressemble un peu au célèbre réseau social Facebook que tu connais peut-être. La principale différence, c'est que Twitter partage des informations entre des personnes qui ont des passions ou des intérêts communs mais qui ne se connaissent pas au départ, alors que Facebook est plutôt destiné à échanger avec des gens que l'on connaît.<br>
#### <font color='red'> Qui utilise Twitter ? </font><br>
On compte 140 millions d'utilisateurs sur toute la planète, dont 5 millions en France ! Parmi eux, il y a aussi des hommes politiques (Nicolas Sarkozy, François Hollande, le président américain Barack Obama…), des stars (Lady Gaga, Justin Bieber…), des entreprises, et des journaux comme le site 1jour1actu !

```python
positive = 0
negative = 0
neutre = 0
for i in twit.tweets:# on va prendre seulement le colonne de texte
    Text_blob= TextBlob(i) # textBlob nous affiche le sentiment
    analyse = Text_blob.sentiment[1]
    if analyse > 0.5:
        positive +=1
    elif analyse < 0.5:
        negative +=1
    else:
        neutre +=1
print("Positives = ", positive)
print("Negatives = ", negative)
print("Neutres = ", neutre)
```

```python
mlk = en_core_web_sm.load() 
Regexp_Tokenizer= RegexpTokenizer(r'\w+')
WordNet_Lemmatizer= WordNetLemmatizer()
stop = set(stopwords.words('english'))
string_punctuation = list(string.punctuation) #already taken care of with the cleaning function.
stop.update(string_punctuation)
Whitespace_Tokenizer= WhitespaceTokenizer()

def fonction(text):
    final_text = []
    for i in Whitespace_Tokenizer.tokenize(text):
        if i.lower() not in stop:
            word = WordNet_Lemmatizer.lemmatize(i)
            final_text.append(word.lower())
    return " ".join(final_text)

twit.tweets = twit.tweets.apply(fonction)
```

```python
def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)

def get_scores(group,tweets):
    scores = []
    for tweets in tweets:
        s = jaccard_similarity(group, tweets)
        scores.append(s)
    return scores

vir_scores = get_scores(Virus, twit.tweets.to_list())
dea_scores = get_scores(death, twit.tweets.to_list())
dis_scores = get_scores(diseases, twit.tweets.to_list())
conf_scores = get_scores(conflict, twit.tweets.to_list())
```

#### <font color=#05144F></font>

```python
# create a jaccard scored df.
date  = {'names':twit.Time.to_list(),       
         'Virus_scores':vir_scores,
         'death_scores': dea_scores, 
         'diseases_scores':dis_scores, 
         'conflict_scores':conf_scores}

scores_df = pd.DataFrame(date)#assign classes based on highest score

def get_classes(l1, l2, l3, l4):
    Viro = []
    deatho = []
    diseaso = []
    conflecto = []
    for i, j, k, l in zip(l1, l2, l3, l4):
        m = max(i, j, k, l)
        if m == i:
            Viro.append(1)
        else:
            Viro.append(0)
        if m == j:
            deatho.append(1)
        else:
            deatho.append(0)        
        if m == k:
            diseaso.append(1)
        else:
            diseaso.append(0)  
        if m == l:
            conflecto.append(1)
        else:
            conflecto.append(0)   
            
    return Viro, deatho, diseaso, conflecto

l1 = scores_df.Virus_scores.to_list()
l2 = scores_df.death_scores.to_list()
l3 = scores_df.diseases_scores.to_list()
l4 = scores_df.conflict_scores.to_list()

Viro, deatho, diseaso, conflecto = get_classes(l1, l2, l3, l4)
date = {'name': scores_df.names.to_list(), 'Virus':Viro, 'death':deatho, 'diseases':diseaso, 'conflict': conflecto}
class_df = pd.DataFrame(date)
#grouping the tweets by username
new_groups_df = class_df.groupby(['name']).sum()#add a new totals column
new_groups_df['total'] = new_groups_df['Virus'] + new_groups_df['death'] + new_groups_df['diseases'] +  new_groups_df['conflict']#add a new totals row
new_groups_df.loc["Total"] = new_groups_df.sum() 
 
```

```python
date = new_groups_df[['conflict', 'Virus','death','diseases']].values

clustering = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.fit(date)
    clustering.append(kmeans.inertia_)
plt.plot(range(1,20), clustering)
plt.xlabel('Number of Clusters')
plt.ylabel('Cluster Sum of Squares')
plt.show()
```
